```python
# Examine data type for INSPECTION_DATE.

df_4['INSPECTION_DATE'].dtype
```

```python
# Convert INSPECTION_DATE data type from object to datetime.

df_4['INSPECTION_DATE'] = pd.to_datetime(df_4['INSPECTION_DATE'])
```

```python
# Check output.

df_4['INSPECTION_DATE'].dtype
```

```python
# Check df_4.

df_4
```

```python
# Extract year from INSPECTION_DATE timestamp. Note: Original data variables are in all caps, derived data variables are in lowercase.

df_4['inspection_year'] = pd.DatetimeIndex(df_4['INSPECTION_DATE']).year
```

```python
# Check output.

df_4
```

```python
# Find earliest inspection_year.

df_4['inspection_year'].min()
```

```python
# Find max inspection_year.

df_4['inspection_year'].max()
```

```python
# Examine data inconsistencies in inspection_year.

df_4['inspection_year'].value_counts(dropna=False)
```

```python
# Create an inspection_year_flag for inspection dates prior to 2000 and after 2024.

result = []

for value in df_4['inspection_year']:
    if value < 2000:
        result.append("Pre-2000")
    elif value > 2024:
        result.append("Impossible")
    else:
        result.append("Present")
```

```python
result
```

```python
# Create new column to contain result.

df_4['inspection_year_flag'] = result
```

```python
# Check value counts for inspection_year_flag.

df_4['inspection_year_flag'].value_counts(dropna = False)
```

```python
# Create dataframe containing only present inspections.

df_5 = df_4[df_4['inspection_year_flag'] == 'Present']
```
